# Exploratory Data Analysis - Multivariate analysis

## Visualization techniques => Multivariate Analysis => **Scatter Plot**

Sometimes we want to know if two variables mean something when put together, whether a small change in one variable affects the other variable. In such cases, plotting a **scatterplot**, or **scatter-diagram**, with our data points can help us to check whether there is a potential relationship between them.

- A **scatterplot** is the simplest mode of a diagrammatic representation of two variables.
- It takes two perpendicular axes of coordinates, one for x and one for y.
- Unlike the lineplot, it directly plots each pair of values as a point on the 2D space.
- The *scatterplot()* function of seaborn can be used to make a scatterplot.

````python
import numpy as np
import pandas as pd

# data visualization
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data/Automobile.csv')
````

````python
df.shape
````

````python
sns.scatterplot(data=df, x = 'engine_size', y = 'horsepower')
plt.show()
````

````python
# we can also check the relationship between two variables for different categories by specifying the hue parameter.
sns.scatterplot(data=df, x = 'engine_size', y = 'horsepower', hue = 'fuel_type', style = 'fuel_type')
plt.show()
````

**Correlation**

Correlation means association. More precisely, it expresses the extent to which two variables change together at a constant rate.

- In a scatter plot when the y variable tends to increase as the x variable increases, we say there is a **positive correlation** between the variables.
- Again, when the y variable tends to decrease as the x variable increases, we say there is a **negative correlation** between the variables.
- If the points on the scatter plot seem to be scattered randomly, we say that there is **no correlation** between the variables.

````python
sns.scatterplot(data=df, x = 'curb_weight', y = 'engine_size')
plt.show()
# positively corelated
````

````python
sns.scatterplot(data=df, x = 'bore', y = 'stroke')
plt.show()
# no co-relation
````

## Visualization techniques => Multivariate Analysis => **Pair Plot**

* A pairplot is a type of data visualization that allows for the exploration of relationships between pairs of variables in a dataset.   
* It creates a grid of scatterplots where each variable is plotted against every other variable, displaying histograms along the diagonal for each variable.
* Pair plots are particularly useful for identifying patterns, correlations, and potential trends between variables, especially in datasets with multiple features.
* The pairplot() function of seaborn can be used to create such a plot.

````python
sns.pairplot(data=df[['normalized_losses', 'wheel_base', 'curb_weight', 'engine_size', 'price', 'peak_rpm']])
````

````python
# we can add the hue parameter in pairplot to create a semantic mapping.
# It changes the default marginal plot to layered kde plot.
# Also, we can add vars parameter to assign a list of variables from the dataset for which we want to create the pairplot.

sns.pairplot(data=df, vars=['normalized_losses', 'wheel_base', 'curb_weight', 'engine_size', 'price', 'peak_rpm'], hue='number_of_doors')
````

## Visualization techniques => Multivariate Analysis => **Heat Plot**
* A heatmap is a data visualization technique that represents data values as colors in a two-dimensional grid.
* It uses a color gradient to encode the magnitude of each data point, with colors ranging from low to high intensity to represent low to high values.
* Heatmaps are commonly ised to visualize patterns, trends, and correlations.


````python
sns.heatmap(data=df[['wheel_base', 'curb_weight', 'engine_size', 'price']].corr())
````

````python
# we can set the annot parameter to True for displaying the numeric calue in each cell.
# To remove the color bar, the cbar parameter can be set to False
sns.heatmap(data=df[['wheel_base', 'curb_weight', 'engine_size', 'price']].corr(), annot=True, cbar=False)
````

````python
# apply a different colormap with the cmap parameter for better visual appeal
sns.heatmap(data=df[['wheel_base', 'curb_weight', 'engine_size', 'price']].corr(), annot=True, cmap='YlGnBu', cbar=False)
````

## Bivariate Analysis => On Melbourne_Housing data
* Bivariate analysis examines the relationship between two variables. It is often represented graphically using scatter plots to visualize the relationship between variables.
* creating bins involves categorizing continuous variables into discrete intervals, or bins, to simplify analysis or visualization

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.float_format', lambda x: '%.3f' % x)
pd.set_option('display.width', 300)

data = pd.read_csv('data/Melbourne_Housing.csv')

data['Date'] = pd.to_datetime(data['Date'], dayfirst=True)
data['BuildingArea'] = data['BuildingArea'].replace(['missing','inf'], np.nan)
data['BuildingArea'] = data['BuildingArea'].astype(float)
data.drop_duplicates(inplace=True)
data.reset_index(drop=True, inplace=True)

data.info()
data.describe().T
````

````python
plt.figure(figsize=(10, 5))
sns.heatmap(data[['Rooms','Date', 'Distance','Postcode', 'Bedroom', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'YearBuilt', 'Propertycount', 'Price']].corr(), annot=True,  cmap='Spectral', vmin=-1, vmax=1)
plt.show()
````

````python
# let's create a column with sum of number of rooms, bedrooms, bathrooms, and car parking spaces
data['Total Space'] = data['Rooms'] + data['Bedroom'] + data['Bathroom'] + data['Car']
data.head()
````

````python
plt.figure(figsize=(10, 5))
sns.scatterplot(data=data, x = 'Total Space', y = 'Price')
````

### lmplot will plot the line of best fit in scatter plot

````python
sns.lmplot(data=data, x = 'Total Space', y = 'Price', height = 5, aspect = 2)
plt.xlim(0, 55)
plt.show()
````

````python
# let's check the correlation between Total Space and Price
data[['Total Space', 'Price']].corr()
````

````python
# let's plot scatter plot between disatnce from the CBD and the price 
plt.figure(figsize=(15,7))
sns.scatterplot(data=data, x = 'Distance', y = 'Price')
plt.show()
````

````python
plt.figure(figsize=(15, 7))
sns.lineplot(data = data, x = 'Distance', y = 'Price', ci=None)
plt.show()
````

### Creating bins for disatnce column
Let's create bins/buckets for the Distance column to get better visualization of the relationship between Price abd Distance
* 0 - 15 Kms => Nearby location
* 15 - 30 KMs => Moderately Close location
* 30 - 50 KMs => Far away location
Use pd.cut() function

``` syntax: pd.cut(x, bins, labels=None, right=False) 

x - column/array to be binned
bins - number of bins to create or an input of list for the range of bins
labels - specifies the labels for the bins
right - If set to False, it excludes the rightmost edge of the interval
```   
The pd.cut() function in Pandas is used to segment and sort data values into discrete intervals, or bins.   
This function is used when dealing with continuous variables that need to be categorized or grouped for further analysis or visualization.

````python
# use pd.cut
data['Distance bins'] = pd.cut(data['Distance'], bins = [0, 15, 30, 50], labels = ['Nearby', 'Moderately Close', 'Far'], right=False)
data.head()
````

````python
sns.boxplot(data=data, x = 'Distance bins', y = 'Price')
plt.show()
````

````python
# Let's not display outliers for easy analysis
sns.boxplot(data=data, x = 'Distance bins', y = 'Price', showfliers=False)
plt.show()
````

## Creating New Series and the use of relplot()
* To create a new series, users can use the pd.series() constructor and pass a list, array, or dictionary of values, along with optional parameters such as index and data type.
* The relplot() function in the Seaborn library is used to create relational plots, which are useful for visualizing the relationship between two variables.

````python
# Analyzing further on the negative co-relation observed between price and year of the property.
year_at_sale = data['Date'].dt.year
year_at_sale
````

````python
data['AgeofProp'] = year_at_sale - data['YearBuilt']
data.head()
````

````python
data[data['AgeofProp'] == -2]
````

````python
# line plot, ci stands for confidence interval. 
# A confidence interval shows the uncertainty around the estimate of the mean. By default, Seaborn shows a 95%
# confidence interval around the line, represented by shaded area around the line.
plt.figure(figsize=(15, 5))
sns.lineplot(data=data, x = 'AgeofProp', y = 'Price', ci=None)
plt.show()
````

## relplot() lets you create multiple plots on a single axis.
kind - specifies the kind of plot to draw (scatter or line)  
ci - soecifies the confidence interval  
col_wrap - specifies the number of columns in the grid  

````python
sns.relplot(data = data, x = 'AgeofProp', y = 'Price', col = 'Regionname', kind = 'line', ci=None, col_wrap=4)
plt.show()
````

````python
# The price of properties vary based on the type of property. For example, a villa may be priced higher than a duplex because of more amenities.

plt.figure(figsize=(10, 5))
sns.boxplot(data = data, x = 'Type', y = 'Price', showfliers = False) # turning off outliers
plt.show()
````

````python
plt.figure(figsize=(10, 5))
sns.boxplot(data = data, x = 'Regionname', y = 'Price', showfliers = False)
plt.xticks(rotation = 90)
plt.show()
````

````python
# Dispersion of price in every region
sns.catplot( x = 'Price',
             col = 'Regionname',
             data = data,
             col_wrap = 4,
             kind = 'violin'
            )
plt.show()
````
