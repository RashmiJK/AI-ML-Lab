## Pandas 
Pandas is a robust and versatile library specifically designed for data analysis in Python. Series, DataFrame, and importing from CSV and Excel files are key features that support machine learning workflows.

```python
import pandas as pd
```

```python
# A pandas **Series** is a one-dimensional labeled array that can hold data of any type (integer, float, string, Python objects, etc.) and provides efficient data manipulation capabilities
members = ["Brazil","Russia","India","China","South Africa"]
brics = pd.Series(members)
print(brics)
print("-" * 20)
type(brics)
```

```python
# A pandas **DataFrame** is a heterogenous two-dimensional data structure with labelled rows and columns. It is similar to a spreadsheet or SQL table, or a dictionary of Series objects.
members = {
    "country":["Brazil","Russia","India","China","South Africa"],
    "capital":["Brasilia","Moscow","New Delhi","Beijing","Pretoria"],
    "gdp": [2750, 1638, 3202, 1409, 416],
    "literacy": [0.92, 0.99, 0.74, 0.96, 0.86],
    "expectancy": [74.56, 70.16, 68.56, 75.38, 57.69],
    "population": [200.4, 143.5, 1252, 1357, 52.98]
}
brics2 = pd.DataFrame(members)
print(brics2)
type(brics2)
```

```python
# Create a DataFrame from previously created two dimesional list of values and a list of column labels
members = [
    ["Brazil","Brasilia",2750,0.92,74.56,200.4],
    ["Russia","Moscow",1638,0.99,70.16,143.5],
    ["India","New Delhi",3202,0.74,68.56,1252],
    ["China","Beijing",1409,0.96,75.38,1357],
    ["South Africa","Pretoria",416,0.86,57.69,52.98]
]
labels = ["country","capital","gdp","literacy","expectancy","population"]
brics3 = pd.DataFrame(members, columns = labels)
print(brics3)
type(brics3)
```

```python
# Data exploration techniques using Pandas features include:
# - .info() - displays DataFrame information like data types and non-null values
# - .describe() - shows summary statistics of numerical columns
# - .head()/.tail() - view first/last n rows
# computing aggregations or statistical measures such as .mean(), .median() and .minimum()
# group-level aggregations
# These help understand data characteristics and summarize key information

washers = pd.read_csv("data/washers.csv")
print(washers.info())
print("--------------------")
print(washers.head())
```

```python
# describe all columns
print(washers.describe())
```

```python
# describing non-numeric columns
print(washers[["BrandName"]].describe())
```

```python
# describing numerical columns
print(washers[["Volume"]].describe())
```

```python
print(washers[['BrandName']].value_counts())
print(washers[['BrandName']].value_counts(normalize = True))
```

```python
print(washers[['Volume']].mean())
```

```python
# Group-level aggregation. Eg: Average volume of washers by brand
print(washers.groupby("BrandName")[["Volume"]].mean())
print(washers.groupby("BrandName")[["Volume"]].mean().sort_values(by="Volume"))
```

```python
# Compute more than one specific aggregation
print(washers.groupby("BrandName")[["Volume"]].agg(["mean","median","min","max"]))
```

## Plot

## A picture is worth a thousand words - Visualization

#### Categorical feature vs Continuous feature
Categorical feature - An attribute that holds data in discrete form, typically limited to a set of possible values. Examples include customer name, grade, purpose, and default.
Continuous feature - An attribute that holds data in the form of an integer or real number, with an infinite number of possible values between its lower and upper bounds. Examples include loan amount, temperature, height, weight, and age.

#### Comparision visualization 
Box plot - Can visualize distribution of continuous feature against the values of categorical feature.
#### Relationship visualization 
Scatter plot, line charts - Can visualize relationship between two or more continuous features.
#### Distribution visualization 
Histogram - Can visualize distribution of the values of a continuous feature. E.g: Most common value of a feature
#### Composition visualization 
Pie chart, stacked bar chart, grouped bar chart - Component makeup of a categorical feature. E.g: How many customers are in each segment

```python
# Use the following magic command to configure Matplotlib to render plots directly within the Jupyter Notebook as static images. 
# This is a built-in feature of Jupyter Notebook. Alternative options include:
# %matplotlib notebook - for interactive plots with zoom and pan functionality within the notebook.
# %matplotlib qt - for interactive plots displayed in a separate window.
%matplotlib inline
```

```python
vehicles = pd.read_csv("data/vehicles.csv")
print(vehicles.info())
print("-" * 20)
print(vehicles.head())
print("-" * 20)
print(vehicles.describe())
```

```python
# Relationship visualization - Scatter plot
vehicles.plot(kind="scatter", x="citympg", y="co2emissions")
```

```python
# Distribution visualization - Histogram
vehicles["co2emissions"].plot(kind="hist")
```

```python
# Comparision visualization - Box plot
vehicles.pivot(columns = "drive", values = "co2emissions")
```

```python
vehicles.pivot(columns = "drive", values = "co2emissions").plot(kind = "box", figsize = (10, 6), title = "CO2 Emissions by Drive Type")
```

```python
# Composition visualization - stacked bar chart
vehicles.groupby("year")["drive"].value_counts()
```

```python
vehicles.groupby("year")["drive"].value_counts().unstack()
```

```python
vehicles.groupby("year")["drive"].value_counts().unstack().plot(kind = "bar", stacked = True, figsize = (10, 6), title = "Drive Type by Year")
```

```python
## Dealing with missing values
import pandas as pd
students = pd.read_excel("data/students.xlsx")
students
```

```python
# List rows with missing values - use isnull method
# First create a mask or filter
mask = students['State'].isnull()
mask
```

```python
# List rows with missing values
# Use the mask to filter the DataFrame
students[mask]
```

```python
# Use dropna() to drop rows with missing values
students.dropna()
```

```python
# Drop rows with missing values for certain columns only
students.dropna(subset = ['State', 'Zip'], how = 'all')
```

```python
# Drop missing value columns
students.dropna(axis = 1)
```

```python
students.dropna(axis = 1, thresh = 10)
```

```python
# fillna() - fill missing values with a specific value
students = students.fillna({'Gender':'Female'})
students
```

```python
students = students.fillna({'Age':students['Age'].median()})
students
```
