```python
import pandas as pd
from pathlib import Path

if Path('./olympics_1896_2004.csv').exists():
    odf = pd.read_csv(filepath_or_buffer='./olympics_1896_2004.csv', skiprows=5)
    print(odf.head())
print("***"*30)
if Path('./INNHotelsGroup_pastdata.csv').exists():
    hdf = pd.read_csv(filepath_or_buffer='./INNHotelsGroup_pastdata.csv')
    print(hdf.head())
```
