## Accessing and modifying DataFrames (loc, iloc)

```python
# Create DataFrame using dictionary
store_data = pd.DataFrame({
    'CustoimerID': ['CustID00','CustID01', 'CustID02', 'CustID03', 'CustID04'],
    'location': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'],
    'gender': ['F', 'M', 'F', 'M', 'F'],
    'type': ['Electornics', 'Food&Beverage', 'Clothing', 'Electornics', 'Food&Beverage'],
    'quantity': [5, 10, 15, 20, 25],
    'total_bill': [500, 1000, 1500, 2000, 2500]
})
print(store_data)
print("\n")
print(store_data.loc[1])
print("\n")
print(store_data.loc[1,'total_bill'])  # Accessing a specific cell using loc
print("\n")
print(store_data.iloc[1, 5])  # Accessing a specific cell using iloc
print("\n")
print(store_data.iloc[1:3, 2:5])  # Accessing a range of rows and columns using iloc
print("\n")
print(store_data.loc[[1,4], ['location', 'type']])  # Accessing specific rows and columns using loc
print("\n")
print(store_data.loc[store_data['total_bill'] > 1500])  
print("\n")
print(store_data.loc[4,'type'])
store_data.loc[4, 'type'] = "Food and Beverages"
print(store_data.loc[4, 'type'])
print("\n")
print(store_data)
```

## Combining dataframes
1. Concatenate using pd.concat()
2. Merge => Merge combines dataframes using a column's values to identify common entries
3. Join => behaves much like merge except instead of using the values of one of the columns to combine data frames, it uses index labels

In Pandas, 'merging out' typically refers to expanding data beyond the initial keys using methods like left, right, or outer joins.

```python
data_cust = pd.DataFrame({
    'customerID': ['101', '102', '103', '104'],
    'category': ['Medium', 'High', 'Low', 'Medium'],
    'first visit': ['yes', 'no', 'yes', 'no'],
    'sales': [1000, 2000, 1500, 3000]}, index=[0, 1, 2, 3])
print(data_cust)
print("\n")

data_cust_new = pd.DataFrame({
    'customerID': ['101', '102', '103', '104'],
    'distance': [12, 9, 15, 10],
    'sales': [1000, 2000, 1500, 3000]}, index=[4, 5, 6, 7])
print(data_cust_new)
print("\n")

print(pd.concat([data_cust, data_cust_new], axis=0))
print("\n")
print(pd.concat([data_cust, data_cust_new], axis=1))
print("\n")
```

```python
data_cust = pd.DataFrame({
    'customerID': ['101', '102', '103', '104'],
    'category': ['Medium', 'High', 'Low', 'Medium'],
    'first visit': ['yes', 'no', 'yes', 'no'],
    'sales': [1000, 2000, 1500, 3000]}, index=[0, 1, 2, 3])
print(data_cust)
print("\n")

data_cust_new = pd.DataFrame({
    'customerID': ['101', '102', '103', '104','105'],
    'distance': [12, 9, 15, 10, 34],
    'sales': [1000, 2000, 1500, 3500, 6000]}, index=[4, 5, 6, 7, 11])
print(data_cust_new)
print("\n")

print(pd.merge(data_cust, data_cust_new, on='customerID', how='outer'))  # Outer join => Union
print("\n")
print(pd.merge(data_cust, data_cust_new, on='customerID', how='inner'))  # Inner join => Intersection
print("\n")
print(pd.merge(data_cust, data_cust_new, on='customerID', how='right'))  # Right join => Right side
```

```python
data_quarters = pd.DataFrame({ 
    'Q1' : [101, 102,103],
    'Q2' : [201, 202, 203]},
    index=['I0', 'I1', 'I2']
)

data_quarters_new = pd.DataFrame({
    'Q3': [301, 302, 303],
    'Q4': [401, 402, 403]
}, index = ['I0', 'I1', 'I2'])

print(data_quarters)
print("\n")
print(data_quarters_new)
print("\n")

print(data_quarters.join(data_quarters_new, how='right'))  # Right join
print("\n")
print(data_quarters.join(data_quarters_new, how='left'))
```

## Pandas functions (groupby)
Used to split the data into groups based on some criteria.
Key differences between `groupby` aggregation and `transform`:
- `transform` returns a DataFrame/Series with the same shape as the input
- `groupby` with aggregation returns reduced results (one row per group)
- `transform` is useful for calculating group-level statistics while maintaining the original data structure

apply() function
sort_values()

The groupby() function in Pandas facilitates the splitting of a DataFrame into groups based on specified criteria, such as unique values in one or more columns, followed by the application of a function to each group, enabling aggregation, tranformation, or analysis of grouped data.

```python
data = pd.read_csv('./data/StockData.csv')
print(data.head())
print("\n")
print(data.groupby(['stock'])['price'].sum())
print("\n")
print(data.groupby(['stock'])['price'].value_counts())
print("\n")
print(data.groupby(['stock'])['price'].transform('sum'))
```

```python
import pandas as pd

# Create sample DataFrame
df = pd.DataFrame({
    'store': ['A', 'A', 'A', 'B', 'B', 'B'],
    'product': ['P1', 'P2', 'P3', 'P1', 'P2', 'P3'],
    'sales': [100, 200, 150, 300, 400, 250]
})

# 1. Simple transform example - Calculate mean sales per store
df['store_mean'] = df.groupby('store')['sales'].transform('mean')
print(df)
print("\n")
print(df.groupby('store')['sales'].transform('min'))
print("\n")
# 2. custom tranform function
def subtract_min(x):
    #print(x)
    return x - x.min()
# Apply custom transform function
df['sales_above_min'] = df.groupby('store')['sales'].transform(subtract_min)
print("\n")
print(df)

# Mean per group
df.groupby('store')['sales'].transform('mean')

# Sum per group
df.groupby('store')['sales'].transform('sum')

# Size of group
df.groupby('store')['sales'].transform('size')

# Rank within group
df.groupby('store')['sales'].transform('rank')
```

```python
# Create more complex DataFrame
sales_df = pd.DataFrame({
    'store': ['A', 'A', 'B', 'B', 'C', 'C'],
    'category': ['food', 'electronics', 'food', 'electronics', 'food', 'electronics'],
    'sales': [100, 200, 150, 300, 250, 400],
    'units': [10, 5, 15, 8, 20, 10]
})

# Multiple transformations
sales_df['sales_ratio'] = sales_df.groupby('store')['sales'].transform(
    lambda x: x / x.mean()
)
sales_df['category_avg'] = sales_df.groupby('category')['sales'].transform('mean')

print("Sales DataFrame with multiple transforms:")
print(sales_df)
```

## Pandas functions (datetime)
Date and time objects can be formatted into readable strings using the **strftime()** method, which allows you to specify the format, such as %y-%m-%d for "Year-Month-Day".
In Pandas, .dt is an accessor to access date and time properties from Series values of datetime64 dtype.

Pandas provide powerful datetime functions for efficient manipulation and analysis of date and time data within DataFrame objects, including parsing, conversion, extraction, and arithmetic operations, enabling comprehensive time-series analysis and data handling.

```python
data = pd.read_csv('./data/StockData.csv')
print(data.sample(5))
print("\n")
data['date'] = pd.to_datetime(data['date'],dayfirst=True)
print(data.info())
print("\n")
print(data.sample(5))
print("\n")
print(data['date'].dt.strftime('%m/%d/%Y'))  
print("\n")
print(data['date'].dt.year)  
print("\n")
print(data['date'].dt.month)
print("\n")
print(data['date'].dt.day)  
print("\n")
print(data['date'].dt.weekday)  
print("\n")
print(data['date'][1] - data['date'][0])
print("\n")
print(data.head(2))
```
